# Kernel Smoothing Method 

Kernel smoothing method is a way of estimating a data point based on its closetes neighbour values. It is different from kernel method, which computes the inner product of high demensional feature space and is used for regularized non linear model. 

This work book serves as a example as how kernel method works.

### K nearest neighbour estimation

```python
K = 10
Est = 4*np.arange(0.0,1,0.05)
# cacluate the distance of the point we need to estimate the data point
import math
def get_near_k(Est,x,y,k):
    # I will accept that
    result = []
    for E in Est:
        dis = np.abs(x - E)
        dis= pd.DataFrame(dis)
        dis.index = y
        dis = dis.sort_values([0],axis=0)
        y_near = dis.index[0:k]
        r = np.mean(y_near)
        result.append(r)
    return result    
```

```python
# You can change the K here and get the shape of x, as K increase to 1000, the fitted line beconme a horizontal line
# Which means assign 
K = 11
result = get_near_k(Est,x,y,K)
plt.scatter(x,y,marker=".",s=size)
plt.scatter(Est,result)
plt.plot(x,np.sin(x))
```

# Nadaraya Waston kernel weighted average

Instead of assign equal weight as the KNN method does Nadaraya Waston Kernel weighted average give those point closer to data higher weight than those data our of range zero weight.

$$ \hat{f(x_0)} = \frac{\sum^N_{i=1}K_{\lambda}(x_{0},x_i)f(x_i)}{\sum^N_{i=1}K_{\lambda}(x_{0},x_i) } $$

$$ K_{\lambda}(x_0,x_i) = D(\frac{|x_0 - x_i |}{\lambda})$$


$$ 
D(t)=\big\{\begin{array}{c l}	
   \frac{3(1-t^2)}{4}   & |t|<=1\\
     0 & otherwise
\end{array} $$

Assume $\lambda = 0.2$, then we could have the fitted value with new kernel

```python
def get_near_k_nadaraya(Est,x,y,lmb):
    # I will accept that
    result = []
    for E in Est:
        dis = np.abs(x - E)
        x_near = dis
        t= x_near/lmb
        a=3*(1-t*t)/4.0
        a[t>1]=0
        r = sum(a*y)/sum(a)
        result.append(r)
    return result   
```

```python
# You can change the K here and get the shape of x, as lambada increase to 1000, the fitted line beconme a horizontal line
# Which means assign 
# This method does not fit well no matter whether the point is too small or too large
lmb = 0.3
result_nada = get_near_k_nadaraya(Est,x,y,lmb)
plt.scatter(x,y,marker=".",s=size)
plt.scatter(Est,result_nada)
plt.plot(x,np.sin(x))
```
